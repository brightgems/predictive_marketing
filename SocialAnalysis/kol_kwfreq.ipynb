```python
df_kwfreq = pd.read_excel("../data/201804/高频词20180301-201831.xls")
# 词过滤
topic_kwds = {
    'Quality Movement':['机芯','机械','机械表','制表','表盘','名表','表壳','智能','表带'],
    'Function/Tech':['自动','计时','技术','防水'],
    'Swiss/Professional':['名表','瑞士','专业','巴塞尔'],
    'sports':['篮球','赛季','球队','NBA','比赛','运动','CBA'],
    'campaign':['圣诞','活动']
}
words = []
for k,v in topic_kwds.items():
    for word in v:
        words.append((k, word))

df_topic_kwds = pd.DataFrame(words, columns= ['主题','高频词'])
df_topicwords = pd.merge(df_kwfreq, df_topic_kwds, on='高频词')
df_topicwords.head()
```
